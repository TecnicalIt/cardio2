```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ccocardio.csv",sep=';')
```

```python
df.head()
```

```python
df["ages"] = (df["age"]/365).round(0)
```

```python
 df["ages"]
```

```python
df.corr()
```

```python
df = df.drop("id",axis=1)
```

```python
df.head()
```

```python
df = df.drop("age",axis=1)
```

```python
df.head()
```

```python
x = df.drop('cardio',axis=1)
```

```python
x.head()
```

```python
y = df.cardio
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.30,random_state=55)
```

```python
xtrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rclf = RandomForestClassifier(n_estimators=100)
```

```python
Rclf.fit(xtrain,ytrain)
```

```python
Rclf.score(xtest,ytest)
```

```python
Rclf.predict([[2169,1,160,80.0,160,100,1,1,0,0,1]])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(xtrain,ytrain)
```

```python
clf.score(xtest,ytest)
```

```python
import pickle as pk
```

```python
with open("my_model",'wb') as file:
    pk.dump(Rclf,file)
```

```python
with open("my_model",'rb') as file:
    my_model1 = pk.load(file)
```
